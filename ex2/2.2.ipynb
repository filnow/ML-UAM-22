```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#indeks 464915
#2.2 A
x = np.arange(0.0, 10.0, 0.01)
y = 5*x**2 - 4*x - 1

plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji")
plt.plot(x, y)
plt.show()
```

```python
#2.2 B
x = np.arange(0.0, 1.5, 0.01)
y_b = np.exp(x)/(np.exp(x)+1)
y_a = 5*x**2 - 4*x - 1

plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji")
plt.plot(x, y_a, color='red')
plt.plot(x, y_b, color='green')
plt.show()
```
